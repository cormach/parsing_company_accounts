```
with open('/annual_reports/Prod223_2701_SC630834_20200404.html') as f:
  example = f.read()
```

```
from bs4 import BeautifulSoup
```

```
soup = BeautifulSoup(example, 'lxml')
```

```
tag_list = soup.find_all()
```

```
for tag in tag_list:
  if tag.has_attr("name"):
    print(tag.attrs,tag.text)
```

```
for tag in tag_list:
  if tag.has_attr("id"):
    print(tag.attrs,tag.text)
```

```
from collections import defaultdict
```

```
tags = defaultdict(set)
for tag in tag_list:
    tags[tag.name] |= set(tag.attrs)

for tag_name in sorted(tags):
    print("{name}:{attrs}".format(name=tag_name, attrs=",".join(sorted(tags[tag_name]))))
```
